## AREA PLOT

### mengelompokkan jumlah kasus tiap wilayah di tahun 2008-2016

```python
df_bor=df['value'].groupby(df['borough']).sum()
df_bor
```

### membuat area plot : Di wilayah mana yang merupakan jumlah kasus tertinggi dan terendah pada tahun 2008-2016?

```python
plt.fill_between(df_bor.index, df_bor.values, alpha=1)
plt.xticks(rotation=90)
plt.xlabel('Borough')
plt.ylabel('Value')
plt.title('Jumlah Kasus Tiap Wilayah pada Tahun 2008-2016', size=14)

plt.show()
```

### Analisis :
Dari area plot diatas, pada tahun 2008-2016 menunjukkan jumlah Kasus tertinggi ada di wilayah 'Westminster' dan jumlah kasus teredah ada di wilayah 'City of London' sedangkan untuk wilayah lain jumlah kasusnya relatif sama

## LINE PLOT

### mengelompokkan jumlah kasus major category di wilayah westminster di tahun 2008-2016

```python
df_west = df[(df['borough']=='Westminster')].groupby('major_category')['value'].sum()
df_west
```

### membuat line plot : Major Category apa yang merupakan kejahatan tertinggi dan terendah di westminster di tahun 2008-2016?

```python
df_west.plot(kind='line',color=['olive'])
plt.xticks(rotation=90)
plt.xlabel('Major Category')
plt.ylabel('Value')
plt.title('Jumlah Kasus Major Category di Wilayah Westminster pada Tahun 2008-2016')

plt.show()
```

### Analisis :
Dari line plot diatas, pada tahun 2008-2016 yang merupakan kasus kejahatan tertinggi (pada ara plot sebelumnya) terjadi di Westminster menunjukkan jumlah Kasus major category tertinggi di Westminster merupakan kejahatan 'Theft and Handling' dan jumlah Kasus major category terendah merupakan kejahatan 'Sexual Offences'

## HISTOGRAM

### mengelompokkan jumlah kasus minor category pada major category theft and handling di wilayah westminster di tahun 2008-2016

```python
df_west_theft = df[(df['borough']=='Westminster')&(df['major_category']=='Theft and Handling')].groupby('minor_category')['value'].sum()
df_west_theft
```

### membuat histogram : kasus minor category apa yang tertinggi dan terendah dalam major category theft and handling di wilayah westminster di tahun 2008-2016?

```python
count,bin_edges = np.histogram(df_west_theft,15)
df_west_theft.plot(kind='hist',figsize=(10,6),bins=15,alpha=0.35,xticks=bin_edges,color=['springgreen'])
plt.xticks(rotation=90)
plt.title('Minor Category dari Theft and Handling di Westminster tahun 2008-2016')
plt.ylabel('Jumlah Minor Category(frequency)')
plt.xlabel('Value')
plt.show()
```

### Analisis : 
dari histogram diatas, sebelumnya berkaitan dengan line plot dimana major category theft handling di westminster merupakan kejahatan terbesar, jika dilihat dari minor category nya maka kejahatan terbesar ada pada 'other theft' dan kejahatan terkecil pada 'handling stolen goods'

## BAR CHART

### Mengelompokkan Total kasus per tahun

```python
df_thn = df.groupby('year')[['value']].sum()
df_thn
```

### membuat bar chart : pada tahun berapa yang merupakan jumlah kasus tertinggi dan terendah?

```python
df_thn.plot(kind='bar',figsize=(10,6),rot=90,color=['salmon'])
plt.title('Total Kasus yang terjadi tiap Tahun')
plt.xlabel('Year')
plt.ylabel('Value')
plt.show()
```

### Analisis :
Dari bar chart diatas terjadi kenaikan dan penurunan relatif sedikit, sehingga bar chart diatas seperti stabil, akan tetapi jumlah Kasus tertinggi pada tahun 2008 dan jumlah kasus teredah pada tahun 2014

## PIE CHART

### Mengelompokkan rata-rata major category tahun 2014

```python
df_mj_14 = df[(df['year']==2014)].groupby('major_category')['value'].mean()
df_mj_14
```

### membuat pie chart : berapa persentase tertinggi dan terendah terhadap major category di tahun 2014?

```python
# ngatur warna dari pie chart
colors_list = ['gold','yellowgreen','lightcoral','lightskyblue','lightgreen','pink','blue','darkgray','beige']
explode_list =[0.1,0,0,0.2,0,0,0.2,0,0] #buat pie chart bisa terlepas dari pie chart total
#urutannya berdasarkan alphabetic
df_mj_14.plot(kind='pie',
              figsize=(15,6),
              autopct='%1.1f%%',
              startangle=90,
              shadow=True,
              labels=None,
              pctdistance=1.12,
              colors=colors_list,
              explode=explode_list)
plt.legend(labels=df_mj_14.index,loc='upper left')
plt.axis('equal')
plt.title('Rata-rata Major Category tahun 2014', y=1)
plt.show()
```

### Analisis :
Dari pie chart diatas, diambil berdasarkan kasus terendah di tahun 2014 pada bar chart sebelumnya, dapat dilihat berdasarkan rata-rata major category di tahun 2014, persentase tertinggi terjadi kasus 'Burglary' dengan persentase 23,6% dan persentase terendah terjadi kasus 'Fraud or Forgery' dan 'Sexual Offences' dengan persentase yang sama yaitu 0%

## BOX PLOT

### mengelompokkan jumlah minor category dari sub burglary tahun 2014 

```python
df_burg = df[(df['year']==2014)&(df['major_category']=='Burglary')].groupby('minor_category')[['value']].sum()
df_burg
```

### membuat box plot : bagaimana minor category dari sub burglary tahun 2014?

```python
df_burg.plot(kind='box',figsize=(8,6))
plt.title('Jumlah Minor Category dari Burglary di tahun 2014')
plt.ylabel('Jumlah Burglary')
plt.show()
```

### Analisis :
* nilai minimal dari jumlah kejahatan burglary sekitar 27000, dan nilai maximal nya sekitar 33000 dan median nya sekitar 38000
* 25% di tahun 2014 memiliki jumlah kejahatan tahunan sebesar -33000 atau kurang (kuartil pertama)
* 75% di tahun 2014 memiliki jumlah kejahatan tahunan sebesar -44000 atau kurang (kuartil kedua)

## SCATTER PLOT

### mengelompokkan jumlah Burglary in a Dwelling di tahun 2008-2016

```python
df_bd = df[(df['minor_category']=='Burglary in a Dwelling')].groupby('year')[['value']].sum()
df_bd.reset_index(inplace=True)
df_bd
```

### membuat scatter plot : pada jumlah Burglary in a Dwelling di tahun 2008-2016, apakah terjadi kenaikan dan/atau penurunan?

```python
df_bd.plot(kind='scatter',x='year',y='value', figsize=(10,6)) #tahunnya makin lama, makin tinggi = korelasi positif
plt.title('Burglary in a Dwelling 2008-2016')
plt.ylabel('total kejahatan')
plt.show()
```

### Analisis :
berdasarkan scatter plot diatas, dapat dilihat bahwa minor category Burglary in a Dwelling dari tahun 2011 sampai 2016 terjadi penurunan kejahatan Burglary in a Dwelling, akan tetapi pada sebelumnya pada tahun 2010 ke 2011 terjadi kenaikan yang cukup signifikan sekitar +5000 kejahatan
